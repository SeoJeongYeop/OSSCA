```python
import numpy as np
import pandas as pd
import os
import json
import datetime
import matplotlib.pyplot as plt
```

```python
statistic_df = pd.read_json('./github_data/github_statistic.json', encoding = 'utf-8', orient='index')
```

```python
def statistic_boxplot(obj, column):
    print("##### ", column, " #####")
    print("sum : ", np.sum(obj))
    print("size: ", np.size(obj))
    print("mean: ", np.mean(obj))
    print("var : ", np.var(obj))
    print("std : ", np.std(obj))
    print("max : ", np.max(obj))
    print("min : ", np.min(obj))
    if(np.max(obj) != 0):
        print("##### max value contributor #####")
        print(obj[obj==np.max(obj)])
    
    plt.boxplot(obj)
    plt.title=column
    plt.show()
```

```python
# 조건걸어서 데이터 추출하는 법
star = statistic_df[statistic_df.stars>0]
star["stars"]
```

```python
for column in statistic_df.columns:
    if(statistic_df[column].dtype == 'O'): # if dtype is Object 
        continue
    statistic_boxplot(statistic_df[column],column)
```
