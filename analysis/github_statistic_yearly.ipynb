```python
def statistic_boxplot(obj, column):
    print("##### ", column, " #####")
    print("sum : ", np.sum(obj))
    print("size: ", np.size(obj))
    print("mean: ", np.mean(obj))
    print("var : ", np.var(obj))
    print("std : ", np.std(obj))
    print("max : ", np.max(obj))
    print("min : ", np.min(obj))
    print("Q1 : ", obj.quantile(.25))
    print("Q2 : ", obj.quantile(.5))
    print("Q3 : ", obj.quantile(.75))
    if(np.max(obj) != 0):
        print("##### max value contributor #####")
        print(obj[obj==np.max(obj)])
    
    plt.boxplot(obj)
    plt.title=column
    plt.show()
```

```python
def checkLen(obj) :
    star = obj[obj.stars==0]
    followers = obj[obj.followers==0]
    followings = obj[obj.followings==0]
    total_repos = obj[obj.total_repos==1]
    total_PRs = obj[obj.total_PRs<4]
    total_issues = obj[obj.total_issues==0]
    print("star: ", len(star))
    print("followers: ", len(followers))
    print("followings: ", len(followings))
    print("total_repos: ", len(total_repos))
    print("total_PRs: ", len(total_PRs))
    print("total_issues: ", len(total_issues))
```

```python
checkLen(statistic_df_2021)
```

```python
# 2021
for column in statistic_df_2021.columns:
    if(statistic_df_2021[column].dtype == 'O'): # if dtype is Object 
        continue
    statistic_boxplot(statistic_df_2021[column],column)
```

```python
for column in score_df_2021.columns:
    if(score_df_2021[column].dtype == 'O'): # if dtype is Object 
        continue
    statistic_boxplot(score_df_2021[column],column)
```
